# Read and clean the data
For start, lets explore what intresting features we have:

We chose to start with the dataset of USA.

```python
df = pd.read_csv("./USvideos.csv")
df.info()
```

The dataset has 40949 videos and 16 features. 

Execept the description feature, all the other are full of values.

Becuase it is only one feature with Null values - we chose to ignore him for our reshearch.

```python
df = df.drop('description', axis=1)
df.head(5)
```

Some of the featuren are numeric and part of them are categorial (for example - category_id).

The category feature is in number which we don't know his meaning.

After read the description in Kaggle website - where we found the dataset: The names of the categories is in the suit Json file for each csv.

So we need to transform the <b>category_id</b> to <b>category_name</b>:

```python
# We will read the Json file and save the appropriate list in the same order as in the csv file
with open("./US_category_id.json") as category_json:
    categories = json.load(category_json)["items"]
category_index_to_name = {}
for catgory in categories:
    category_index_to_name[int(catgory["id"])] = catgory["snippet"]["title"]
df['category_name'] = df['category_id'].map(category_index_to_name)
df = df.drop('category_id', axis=1)
```

```python
df.head(5)
```

```python
df.category_name.unique()
```

Now the categories are easy to understand.

Also we will want to reshearch the time for publish a video, so we want to saperate the feature <b>publish_time</b> for: publish_date and publish_hour:

```python
# Get the publish month for each video
df['publish_date'] = df['publish_time'].map(lambda date: datetime.strftime(datetime.strptime(date, '%Y-%m-%dT%H:%M:%S.000Z'), '%d-%m-%Y'))
df['publish_hour'] = df['publish_time'].map(lambda date: datetime.strftime(datetime.strptime(date, '%Y-%m-%dT%H:%M:%S.000Z'), '%H:%M:%S'))
df = df.drop('publish_time', axis=1)
df.head(5)
```

```python
#We transform the feature trending_date to the Day-Month-Year frame
df['trending_date'] = df['trending_date'].map(lambda date: datetime.strftime(datetime.strptime(date, '%y.%d.%m'), '%d-%m-%Y'))
df.head(5)
```

## The number of videos follow the years - Usage of YouTube

We want to learn if the number of trending videos is from uniform distribution or is there a peroid where more videos publish then?

We chose to show it with histogram becuase it is the simple way to show the distribution over a lot of months.

```python
df['publish_month'] = df['publish_date'].map(lambda date: datetime.strftime(datetime.strptime(date, '%d-%m-%Y'), '%Y-%m'))
count_per_day = df.loc[df['publish_month'] >= '2010-01'].groupby(['publish_month']).size().reset_index(name='count')
# Plot graph
sns.set(rc={'figure.figsize':(15.7,5.27)})
plt.grid(None)
sns.barplot(x='publish_month', y='count', data=count_per_day)
t = plt.xticks(rotation=90)
plt.title('Count of videos across time', size=14)
```

We can see the increase of records in the dataset is between 11-2017 till 05-2018.

Between those months there is no dramatic increase or decrease we can learn from it.

So we will cut all the videos who aren't publish till the January 2017.

We define that - all our assumptions and conclusions are for those 2 years.

```python
df = df[df['publish_month']>='2017-01']
count_per_day = df.loc[df['publish_month'] >= '2010-01'].groupby(['publish_month']).size().reset_index(name='count')
# Plot graph
print('Now the dataset has '+str(df.shape[0])+' videos.')
print('That means we remove '+str(40949-df.shape[0])+' videos.')
sns.set(rc={'figure.figsize':(15.7,5.27)})
plt.grid(None)
sns.barplot(x='publish_month', y='count', data=count_per_day)
t = plt.xticks(rotation=90)
plt.title('Count of videos across time', size=14)
# remove the tmp columns:
df = df.drop('publish_month',axis=1)
```

## Publish_time

For each video we have his publish date. We think maybe if video publish in weekend or workday it has an affect to his success?

We choose histogram becuase it is the simplest way see amount of videos for each possible value.

```python
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]

def calculate_day(input_date):
    day, month, year = (int(x) for x in input_date.split('-'))    
    input_data = date(year, month, day)
    dayNumber=input_data.weekday()
    return days[dayNumber]

df['day'] = [calculate_day(date) for date in df['publish_date']]
days_df = df.groupby('day').agg({'video_id':'count', 'views': 'sum'}).reset_index().rename(columns={'video_id':'video_count', 'views': 'total_views'})
plt.grid(None)
sns.barplot(x="day", y="video_count", data=days_df, order=days)
plt.xlabel('day')
plt.ylabel('count of videos')
plt.xticks(rotation=90)
plt.title('The distribution for publish day', size=16)
plt.show()
```

We can see the saparate between the workday to weekend. In each sub-group of days the count of videos is almost the same.

In the workday there is larges amount of videos that publish than in the weekend.

According to this result, it will make easy to see the relation between the ratio between views per video in day publish and proportion likes per views and comments per views:

```python
days_df['prop_views'] = np.divide(days_df['total_views'], days_df['video_count'])
plt.grid(None)
sns.barplot(x="day", y="prop_views", data=days_df, order=days)
plt.xlabel('day')
plt.ylabel('count of views for video in publish day')
plt.xticks(rotation=90)
plt.title('The distribution for views per video in publish day', size=16)
plt.show()
```

Here we can see another point of view:
In the workday there isn't a big change in the views per video, but when it's get closer to weekend - there is an increase in the amount of views per video: specially in Friday- the start of the weekend.

We expect to see also here a uniformally saperation in the views in the weekend, but there is a big decrease in the amount of views in the videos which were publish in Saturday.

We can see that most popular videos are the ones which were publish in Friday and Sunday - the weekend.

Probably there is a connection to the weekend: it is hard to get specific conclusion about it becuase we don't have information about how really the views are saparated over time isn't publish to us, maybe the uniformally assumption isn't work here.

We choose box plot - beacuase we want to see the distribution for each value sapartely.

```python
plt.grid(None)
sns.boxplot(x="day", y="likes_views_prop", data=df, order=days)
plt.xticks(rotation=90)
plt.title('The distribution for likes per views in publish day', size=16)
plt.show()
```

Surprisingly we don't see a diffrance between the days in this feature.

But with the conclusions we get in the previos graphs we can say that becuase it is a proportion and we know that most of the videos are publish in the workday have less views - we can assume that they have also low amount of likes. <font size="6"><font color="red"><b>Can we explore state habits in the internet?</b></font></font>

Another question we ask in our project suggestion is to explore the countries habits in the web.

Can we learn diffrance or similarity between countries?

This dataset has csv file for each country sapartly.
For analysis we union them to one dataset and explore their intersts and habits.

```python
country_list = ['CA', 'DE', 'FR', 'GB', 'US']
countries_full_names = ['Canada', 'Germany', 'France', 'Great Britian', 'USA']
union_df = pd.DataFrame()
for index,country in enumerate(country_list):
    # Retreive the category names and insert them to the dataframe
    print
    with open("./" + country + "_category_id.json") as category_json:
        categories = json.load(category_json)["items"]
    category_index_to_name = {}
    for catgory in categories:
        category_index_to_name[int(catgory["id"])] = catgory["snippet"]["title"]

    df_tmp = pd.read_csv("./" + country + "videos.csv")
    df_tmp['category_name'] = df_tmp['category_id'].map(category_index_to_name)
    country_full_name = countries_full_names[index]
    df_tmp['country'] = [country_full_name for x in range(df_tmp.shape[0])]
    union_df = pd.concat([union_df,df_tmp])

union_df.head(10)
```

We assume that the dataset does present the population of all videos in YouTube.

We define our hypotheses:

### Null hypothese: There is no different in average of views between official video or not in YouTube

### Alternative hypothese: There is a differance in average of views between official video or not in YouTube

Our satistics parameter is the average of views for all videos.

We assume that our dataset ir representing the trending videos in YouTube - so we will use the bootstrap method.

We can calculate one value of the test statistic under the null hypothesis. We shuffle our feature if there is official in the title and check if the average of views has changed? 

We can therefore create an empirical distribution fot the statistic, compare to the observed value and find a p-value with reference to the original official video's proportion in each category.

## Hypotheses Test

```python
def diff_of_avgs(df):
    grpby_var = df.groupby('is_title_include')
    avgs = grpby_var['views'].mean()
    return avgs[1] - avgs[0]


def create_shuffle_dataset_reference_categories(dict_data, prop_dict, list_cat_names):
    new_df = pd.DataFrame(columns=['views', 'is_title_include'], dtype=float)
    for category in list_cat_names:
        amount_for_official_videos = prop_dict[category]
        tmp_df = dict_data[category]
        df_sample = tmp_df.sample(amount_for_official_videos, replace=False)
        tmp_df['is_title_include'] = [1 if video[0] in df_sample.index else 0 for video in tmp_df.iterrows()]
        new_df = new_df.append(tmp_df, ignore_index=True)
    return new_df


def simulations_mean_difference(dict_data, prop_dict, list_cat_names, num_replications):
    res_mean_diffs = np.empty(num_replications)
    for i in range(num_replications):
        one_sample = create_shuffle_dataset_reference_categories(dict_data, prop_dict, list_cat_names)
        resampled_mean_diff = diff_of_avgs(one_sample)
        res_mean_diffs[i] = resampled_mean_diff
    return res_mean_diffs


data = df[['video_id','views','category_name','is_title_include']]
observed_diff = diff_of_avgs(data)
categories_names = list(set(data['category_name']))
dict_dataframes = {}
ori_count_official_per_category = {}
for category in categories_names:
    tmp_df = data[data['category_name']==category]
    count_official = tmp_df[tmp_df['is_title_include'] == 1].shape[0]
    ori_count_official_per_category[category] = count_official
    dict_dataframes[category] = tmp_df[['views','is_title_include']]

# Run the shuffle 2000 times
res_simulations = simulations_mean_difference(dict_dataframes, ori_count_official_per_category, categories_names, 2000)

# Calculate P-Value
num_simulations_like_data_or_more_extreme = np.count_nonzero(res_simulations >= observed_diff) 
print ('The p-value is', num_simulations_like_data_or_more_extreme/2000)

# Get the endpoints of the 95% confidence interval
left_end = np.percentile(res_simulations, 2.5, interpolation='higher')
right_end = np.percentile(res_simulations, 97.5, interpolation='higher')
print('The 95% boostsrap confidence interval for difference between population means', [left_end,right_end])

# visualize results
plt.grid(None)
ax = sns.distplot(res_simulations, kde=False)
ax.set(xlabel='diff of averages views', ylabel='count simulations')
ax.set_title('count of simulations per value of diffrance average of views')
plt.hlines(y=0, xmin=left_end, xmax=right_end, colors='orange', linestyles='solid', lw=4)
plt.scatter(observed_diff, 0, color='red', s=30)
```

According the P-Value test, we got 0 as it - so for each positive statistical significance (and of course for 0.05), we can deny our null hypothesis.

Also with the confidence interval test- we can deny our null hypothesis, 0 isn't belong to the empirial given confidence interval.

## Conclusion: There is a diffrence in average of views in videos with 'official' in the title or not.

Becuase we want to define this faeture as the response parameter - we need to transform it to categorial.

So we need to find the quantile for 1/3 and 2/3 - the values who will separate the values to low, medium, high:

```python
quantile_third = math.ceil(np.quantile(df['views'], 1/3))
quantile_two_third = math.ceil(np.quantile(df['views'], 2/3))
print('The quantile third is: '+str(quantile_third))
print('the quantile two third is: '+str(quantile_two_third))
```

```python
def categorize_views(num_views):
    if num_views <= quantile_third:
        return 'Low'
    elif quantile_third < num_views <= quantile_two_third:
        return 'Medium'
    else:
        return 'High'
df['views_by_catagory']=df['views'].apply(categorize_views)
```

In part 1 - we fount the connection between the views feature to likes and comment_count parameter, so we use them as the parameters.

We see that the likes and comment_count are in diffrent scales - so we will norm then in the same scale.

We chose with the min-max scale.

```python
def scale_column_by_min_max(df_col):
    scaled_df_col = (df_col - df_col.min()) / (df_col.max() - df_col.min())
    return scaled_df_col
df_new=df.copy()
#  We need to scale the values of the predictors
df_new['likes_scaled'] = scale_column_by_min_max(df_new['likes'])
df_new['comment_count_scaled'] = scale_column_by_min_max(df_new['comment_count'])
display(df_new)
cx = sns.relplot(kind='scatter', x='likes_scaled', y='comment_count_scaled', hue='views_by_catagory', data=df_new)
plt.title('comment count scaled per liked scaled with views classification')
plt.grid(None)
cx.set_xticklabels(rotation=45)
```

Now we want to find the optimal K as number of neighbors.

We choose to show it as a linear graph with dots becuase to hopfully get a defined connection between the grade to parameter k.

```python
df_shuffle = df.sample(frac=1)
X=df_new.loc[:,['likes_scaled','comment_count_scaled']].values
Y=df_new.loc[:,'views_by_catagory'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)

# Create a new KNN model
scores_list=[]
scoring = ["f1_macro", "accuracy"]
k_list = range(1, 26)
for k in k_list:
    knn_cv = KNeighborsClassifier(n_neighbors=k)
    cv_scores = cross_validate(knn_cv, X_train, Y_train ,cv=5, scoring=scoring)
    scores_list.append(cv_scores)

f1_scores = [np.mean(score['test_f1_macro']) for score in scores_list]
accuracy_scores = [score['test_accuracy'] for score in scores_list]
mean_score_list = list(map(np.mean, accuracy_scores))
std_score_list = list(map(np.std, accuracy_scores))
```
